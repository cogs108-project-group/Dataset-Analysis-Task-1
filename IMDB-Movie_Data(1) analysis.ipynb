```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_data=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
movie_data
```

```python
desc= movie_data.describe()
desc
```

```python
plt.hist(x=movie_data['Year'], bins =20)
plt.xlim(2006,2016)
plt.xlabel('year')
plt.ylabel('count')
```

```python
plt.hist(x=movie_data['Revenue (Millions)'], bins=50)
plt.xlabel('Revenue in millions')
plt.ylabel("count")

```

```python
plt.scatter(x=movie_data['Revenue (Millions)'], y=movie_data['Year'], s=30)
plt.xlabel('Revenue in millions')
plt.ylabel("Year")
```

```python
plt.scatter(x=movie_data['Revenue (Millions)'], y=movie_data['Runtime (Minutes)'], s=30)
plt.xlabel('Revenue in millions')
plt.ylabel("Runtime")
```
